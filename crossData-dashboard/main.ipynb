```python
import dask.dataframe as dd
import pandas as pd
import numpy as np
import streamlit as st
```

```python
# Datasets do Candi
candiSentimentos = dd.read_csv("candiSentimentos.csv").compute()
candiSintomas = dd.read_csv("candiSintomas.csv").compute()

# ______________________________ #
# Datasets encontrados no Kaggle #
# ------------------------------ #

# SOBRE: dados cientificos sobre o cancer (tumores etc)
datasetSerio = dd.read_csv("dataSeria.csv").compute()

# SOBRE: informacoes gerais sobre o cancer no SUS
datasetSUS = dd.read_csv(
    "datasetSUS.csv",
    dtype={
        'ANOMES_TRA': 'float64',
        'CNES_TRAT': 'float64',
        'MUN_TRATAM': 'float64',
        'UF_TRATAM': 'float64',
    }
).compute()

# SOBRE: dados estatisticos sobre o cancer na midia jornalistica
datasetNoticias = dd.read_csv("noticiasCancer.csv").compute()

# SOBRE: dados sobre os sentimentos de pacientes com cancer (comentarios similar ao Candi!!)
datasetSentimentos2 = dd.read_csv(
    "sentimentosOncologia.csv",
    sep=";",
    encoding="utf-8-sig",
    dtype={'none': 'object'}
).compute()

# SOBRE: dados sobre as taxas de sobrevivencia de pacientes com cancer
datasetSobrevivencia = dd.read_csv("sobrevivenciaCancer.csv").compute()

# SOBRE: dados sobre o tempo para o inicio do tratamento de pacientes com cancer no Brasil
datasetTempoTratamento = dd.read_csv(
    "tempoP_inicioTratamento.csv",
    encoding="latin-1"  
).compute()
```
